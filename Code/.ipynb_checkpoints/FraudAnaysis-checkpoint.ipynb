# EDA Further

## Cleaning data(Outliers)

Outliers only exist in one of our columns that is total_cost_to_total_charges_ratio.

```python
insurance_data["ratio_of_total_costs_to_total_charges"]
```

```python
from statsmodels.graphics.gofplots import qqplot
```

```python
qqplot(insurance_data["ratio_of_total_costs_to_total_charges"], line = 's', fit = True)
```

Calculate Q1 and Q3 for checking outliers

```python
Q1, Q3 = insurance_data["ratio_of_total_costs_to_total_charges"].quantile([0.25, 0.75])
```

```python
IQR = Q3 - Q1

lower_limit = Q1 - (1.5 * IQR)
upper_limit = Q3 + (1.5 * IQR)
```

```python
print(f"Q1: {Q1}\nQ3: {Q3}\nIQR: {IQR}\nlower_limit: {lower_limit}\nupper_limit: {upper_limit}")
```

making list of outliers, also those outliers which are smaller than Q1 and one list for outliers bigger than Q3

```python
outliers = []
outliers_min = []
outliers_max = []
for element in insurance_data["ratio_of_total_costs_to_total_charges"]:
    if element < lower_limit:
        outliers.append(element)
        outliers_min.append(element)
    elif element > upper_limit:
        outliers.append(element)
        outliers_max.append(element)
```

```python
print(f"Total number of outliers: {len(outliers)}")
print(f"Outliers below Lower Limit: {len(outliers_min)}")
print(f"Outliers above Upper Limit: {len(outliers_max)}")
```

We have total 64576 outliers. If we remove them it would make a greater impact so removing outlirs will not be a good option

We can use capping means assigning using Q3 for upper outliers and Q1 for lower outliers. or We can impute all the outliers with mean or median

## Handling outliers

### Creating Function

```python
# This function will handle outliers by imputing new value(median) or use capping
def remove_outliers(data, method):
    new_values = []
    
    if method == "imputation":
        median = data.median()
        
        for element in data:
            
            if (element < lower_limit) or (element > upper_limit):
                new_values.append(median)
                
            else:
                new_values.append(element)
    
    elif method == "capping":
        q1, q3 = data.quantile([0.25, 0.75])
        
        for element in data:
            
            if (element < lower_limit):
                new_values.append(q1)
                
            elif (element > upper_limit):
                new_values.append(q3)
                
            else:
                new_values.append(element)
                
    return new_values
```

```python
X = undersample_data.drop(columns = ["Result"])
y = undersample_data["Result"]

Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size = .25)
```

```python
models_accuracies = []
```

```python
def ModelBuilding(n_ests, crit):
    random_forest = RandomForestClassifier(n_estimators = n_ests, criterion = crit, n_jobs = -1)
    model = random_forest.fit(Xtrain, ytrain)
    scores = cross_validate(model, Xtrain, ytrain, n_jobs = -1, cv = 6, scoring = 'f1_micro')
    train_score = round(scores["test_score"][0], 4)
    ypred = model.predict(Xtest)
    test_score = round(accuracy_score(ytest, ypred), 4)

    current_model = {
        'model': "Random Forest",
        'n_estimators': n_ests,
        'criterion': crit,
        'train_score': train_score,
        'test_score': test_score
    }

    models_accuracies.append(current_model)
```

```python
import random
```

```python
criterion = ["gini", "entropy"]
for i in range(50, 401, 50):
    random_criterion = random.randint(0, 1)
    ModelBuilding(i, criterion[random_criterion])
```
